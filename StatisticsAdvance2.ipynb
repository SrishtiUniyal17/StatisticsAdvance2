```python
# Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution

def bernoulli_pdf(x, p):
    """
    Calculates the probability distribution function (PDF) of a Bernoulli distribution.

    Parameters:
    x (int): The outcome (0 or 1)
    p (float): The probability of success (x = 1), 0 <= p <= 1

    Returns:
    float: The probability of the outcome x
    """
    if x not in (0, 1):
        raise ValueError("x must be 0 or 1 for a Bernoulli distribution.")
    if not (0 <= p <= 1):
        raise ValueError("Probability p must be between 0 and 1.")

    return p if x == 1 else 1 - p

# Example usage
print("P(X=0) when p=0.3:", bernoulli_pdf(0, 0.3))
print("P(X=1) when p=0.3:", bernoulli_pdf(1, 0.3))
```

```python
# Write a Python function to calculate Z-scores from a dataset and plot them

import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    """
    Calculates the Z-scores for a dataset.

    Parameters:
    data (array-like): List or NumPy array of numerical values.

    Returns:
    np.ndarray: Z-scores corresponding to the data.
    """
    data = np.asarray(data)
    mean = np.mean(data)
    std_dev = np.std(data)

    if std_dev == 0:
        raise ValueError("Standard deviation is zero. Z-scores are undefined.")

    z_scores = (data - mean) / std_dev
    return z_scores

def plot_z_scores(z_scores):
    """
    Plots Z-scores as a scatter plot and histogram.

    Parameters:
    z_scores (array-like): Computed Z-scores to plot.
    """
    plt.figure(figsize=(12, 5))

    # Scatter plot
    plt.subplot(1, 2, 1)
    plt.scatter(range(len(z_scores)), z_scores, alpha=0.7, color='blue')
    plt.axhline(0, color='red', linestyle='--', label='Mean (Z=0)')
    plt.title("Z-Scores (Scatter Plot)")
    plt.xlabel("Index")
    plt.ylabel("Z-Score")
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.legend()

    # Histogram
    plt.subplot(1, 2, 2)
    plt.hist(z_scores, bins=30, density=True, alpha=0.6, color='green', edgecolor='black')
    plt.title("Z-Scores Distribution (Histogram)")
    plt.xlabel("Z-Score")
    plt.ylabel("Density")
    plt.grid(True, linestyle='--', alpha=0.5)

    plt.tight_layout()
    plt.show()

# Example usage
if __name__ == "__main__":
    # Simulate sample data
    sample_data = np.random.normal(loc=50, scale=10, size=100)

    # Calculate Z-scores
    z = calculate_z_scores(sample_data)

    # Plot Z-scores
    plot_z_scores(z)
```

```python
''' Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
distribution '''

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(data_point, mean, std_dev):
    """
    Calculate the Z-score for a given data point.

    Parameters:
    data_point (float): The data point for which to calculate the Z-score.
    mean (float): The mean of the dataset.
    std_dev (float): The standard deviation of the dataset.

    Returns:
    float: The Z-score for the given data point.
    """
    return (data_point - mean) / std_dev

def compare_z_score_with_normal(z_score):
    """
    Compare the Z-score to the standard normal distribution and calculate the probability
    using the CDF of the standard normal distribution.

    Parameters:
    z_score (float): The Z-score to compare with the standard normal distribution.

    Returns:
    float: The probability (CDF) corresponding to the Z-score in the standard normal distribution.
    """
    # CDF value for the Z-score (the probability of getting a value <= z_score in a standard normal distribution)
    probability = norm.cdf(z_score)
    return probability

# Example data
data_point = 75  # The data point for which to calculate the Z-score
data = np.random.normal(loc=70, scale=10, size=1000)  # Example dataset (normal distribution)

# Calculate mean and standard deviation of the dataset
mean = np.mean(data)
std_dev = np.std(data)

# Calculate Z-score for the data point
z_score = calculate_z_score(data_point, mean, std_dev)

# Compare the Z-score with the standard normal distribution
probability = compare_z_score_with_normal(z_score)

# Display the results
print(f"Data Point: {data_point}")
print(f"Mean of the dataset: {mean:.2f}")
print(f"Standard Deviation of the dataset: {std_dev:.2f}")
print(f"Z-Score: {z_score:.2f}")
print(f"Probability from the standard normal distribution (CDF): {probability:.4f}")

# Plot the standard normal distribution and the Z-score
x = np.linspace(-4, 4, 1000)
pdf = norm.pdf(x, 0, 1)  # Standard normal distribution

plt.figure(figsize=(8, 5))
plt.plot(x, pdf, label="Standard Normal Distribution (mean=0, std=1)")
plt.axvline(x=z_score, color='red', linestyle='--', label=f"Z-Score: {z_score:.2f}")
plt.fill_between(x, pdf, where=(x >= z_score), color='red', alpha=0.3)

# Add labels and title
plt.title("Standard Normal Distribution and Z-Score")
plt.xlabel("Z-Score")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# Implement hypothesis testing using Z-statistics for a sample dataset

import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

def z_test(sample_data, population_mean, population_std, alpha=0.05):
    """
    Perform a one-sample Z-test to test if the sample mean differs from the population mean.

    Parameters:
    sample_data (array-like): The sample data.
    population_mean (float): The mean of the population under the null hypothesis.
    population_std (float): The standard deviation of the population (known).
    alpha (float): The significance level (default is 0.05).

    Returns:
    None: Prints results of the hypothesis test.
    """
    # Calculate the sample mean and sample size
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)

    # Calculate the Z-statistic
    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Calculate the critical Z-value for the two-tailed test at the given alpha level
    z_critical = norm.ppf(1 - alpha / 2)  # Two-tailed test, 1 - alpha/2

    # Calculate the p-value for the test
    p_value = 2 * (1 - norm.cdf(abs(z_statistic)))  # Two-tailed p-value

    # Display the results
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-statistic: {z_statistic:.2f}")
    print(f"Critical Z-value: {z_critical:.2f}")
    print(f"P-value: {p_value:.4f}")

    if abs(z_statistic) > z_critical:
        print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
    else:
        print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")

    # Visualize the test
    visualize_z_test(z_statistic, z_critical)

def visualize_z_test(z_statistic, z_critical):
    """
    Visualize the Z-test and mark the Z-statistic and critical value.

    Parameters:
    z_statistic (float): The calculated Z-statistic.
    z_critical (float): The critical Z-value at the given alpha level.

    Returns:
    None: Displays the plot.
    """
    # Plot the standard normal distribution
    x = np.linspace(-4, 4, 1000)
    pdf = norm.pdf(x, 0, 1)

    plt.figure(figsize=(8, 5))
    plt.plot(x, pdf, label="Standard Normal Distribution", color="blue")

    # Shade the critical region (outside the critical Z-values)
    plt.fill_between(x, pdf, where=(x >= z_critical) | (x <= -z_critical), color='red', alpha=0.3, label="Rejection Region")

    # Mark the Z-statistic and critical values
    plt.axvline(z_statistic, color='green', linestyle='--', label=f"Z-statistic: {z_statistic:.2f}")
    plt.axvline(z_critical, color='orange', linestyle='--', label=f"Critical Z-value: ±{z_critical:.2f}")

    plt.title("Z-Test Visualization")
    plt.xlabel("Z-value")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()

# Example usage
if __name__ == "__main__":
    # Sample data (e.g., test scores, height, etc.)
    sample_data = np.random.normal(loc=52, scale=10, size=100)  # Sample of 100 data points

    # Population parameters
    population_mean = 50  # Hypothetical population mean
    population_std = 10   # Population standard deviation (known)

    # Perform Z-test
    z_test(sample_data, population_mean, population_std, alpha=0.05)
```

```python
# Create a confidence interval for a dataset using Python and interpret the result

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def calculate_confidence_interval(data, confidence_level=0.95):
    """
    Calculate the confidence interval for a given dataset.

    Parameters:
    data (array-like): The sample data.
    confidence_level (float): The confidence level (default is 95%).

    Returns:
    tuple: The lower and upper bounds of the confidence interval.
    """
    # Sample statistics
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Sample standard deviation (Bessel correction)
    sample_size = len(data)

    # Standard error
    standard_error = sample_std / np.sqrt(sample_size)

    # Find the t-score for the given confidence level
    degrees_of_freedom = sample_size - 1
    t_critical = stats.t.ppf((1 + confidence_level) / 2, df=degrees_of_freedom)

    # Calculate the margin of error
    margin_of_error = t_critical * standard_error

    # Calculate the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

def plot_confidence_interval(data, confidence_interval):
    """
    Plot the sample data along with the confidence interval.

    Parameters:
    data (array-like): The sample data.
    confidence_interval (tuple): The confidence interval (lower bound, upper bound).

    Returns:
    None: Displays the plot.
    """
    # Plot the sample data (histogram)
    plt.figure(figsize=(8, 5))
    plt.hist(data, bins=20, color='skyblue', edgecolor='black', alpha=0.7, label="Sample Data")

    # Plot the confidence interval
    plt.axvline(confidence_interval[0], color='red', linestyle='--', label=f"Lower Bound: {confidence_interval[0]:.2f}")
    plt.axvline(confidence_interval[1], color='green', linestyle='--', label=f"Upper Bound: {confidence_interval[1]:.2f}")
    plt.axvline(np.mean(data), color='black', linestyle='-', label=f"Sample Mean: {np.mean(data):.2f}")

    # Add labels and title
    plt.title(f"Confidence Interval for Sample Data (Confidence Level: 95%)")
    plt.xlabel("Data Value")
    plt.ylabel("Frequency")
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()

# Example usage
if __name__ == "__main__":
    # Example data: A sample of test scores
    data = np.random.normal(loc=75, scale=10, size=100)  # Normally distributed data with mean=75, std=10

    # Calculate 95% confidence interval
    confidence_interval = calculate_confidence_interval(data, confidence_level=0.95)

    # Display the result
    print(f"Sample Mean: {np.mean(data):.2f}")
    print(f"95% Confidence Interval: ({confidence_interval[0]:.2f}, {confidence_interval[1]:.2f})")

    # Plot the confidence interval
    plot_confidence_interval(data, confidence_interval)
```

```python
# Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def calculate_confidence_interval(data, confidence_level=0.95):
    """
    Calculate the confidence interval for a given dataset.

    Parameters:
    data (array-like): The sample data.
    confidence_level (float): The confidence level (default is 95%).

    Returns:
    tuple: The lower and upper bounds of the confidence interval.
    """
    # Sample statistics
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Sample standard deviation (Bessel correction)
    sample_size = len(data)

    # Standard error
    standard_error = sample_std / np.sqrt(sample_size)

    # Find the t-score for the given confidence level
    degrees_of_freedom = sample_size - 1
    t_critical = stats.t.ppf((1 + confidence_level) / 2, df=degrees_of_freedom)

    # Calculate the margin of error
    margin_of_error = t_critical * standard_error

    # Calculate the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

def plot_confidence_interval(data, confidence_interval):
    """
    Plot the sample data along with the confidence interval.

    Parameters:
    data (array-like): The sample data.
    confidence_interval (tuple): The confidence interval (lower bound, upper bound).

    Returns:
    None: Displays the plot.
    """
    # Plot the sample data (histogram)
    plt.figure(figsize=(8, 5))
    plt.hist(data, bins=20, color='skyblue', edgecolor='black', alpha=0.7, label="Sample Data")

    # Plot the confidence interval
    plt.axvline(confidence_interval[0], color='red', linestyle='--', label=f"Lower Bound: {confidence_interval[0]:.2f}")
    plt.axvline(confidence_interval[1], color='green', linestyle='--', label=f"Upper Bound: {confidence_interval[1]:.2f}")
    plt.axvline(np.mean(data), color='black', linestyle='-', label=f"Sample Mean: {np.mean(data):.2f}")

    # Add labels and title
    plt.title(f"Confidence Interval for Sample Data (Confidence Level: 95%)")
    plt.xlabel("Data Value")
    plt.ylabel("Frequency")
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()

# Example usage
if __name__ == "__main__":
    # Generate data from a normal distribution (mean=100, std=15)
    np.random.seed(0)  # For reproducibility
    data = np.random.normal(loc=100, scale=15, size=50)  # 50 data points from normal distribution

    # Calculate 95% confidence interval
    confidence_interval = calculate_confidence_interval(data, confidence_level=0.95)

    # Display the result
    print(f"Sample Mean: {np.mean(data):.2f}")
    print(f"95% Confidence Interval: ({confidence_interval[0]:.2f}, {confidence_interval[1]:.2f})")

    # Plot the confidence interval
    plot_confidence_interval(data, confidence_interval)
```

```python
''' Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results '''


import numpy as np
from scipy.stats import norm

def z_test(sample, population_mean, population_std, alpha=0.05):
    """
    Perform a Z-test for comparing a sample mean to a known population mean.

    Parameters:
    sample (array-like): The sample data.
    population_mean (float): The population mean.
    population_std (float): The population standard deviation.
    alpha (float): The significance level (default is 0.05).

    Returns:
    tuple: Z-statistic, critical value, decision (Reject or Fail to Reject H0)
    """
    # Calculate sample mean and sample size
    sample_mean = np.mean(sample)
    n = len(sample)

    # Calculate the Z-statistic
    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    # Find the critical value from the standard normal distribution
    critical_value = norm.ppf(1 - alpha / 2)  # Two-tailed test

    # Compare Z-statistic to critical value
    if abs(z_statistic) > critical_value:
        decision = "Reject the null hypothesis (H0)"
    else:
        decision = "Fail to reject the null hypothesis (H0)"

    return z_statistic, critical_value, decision

# Example usage
if __name__ == "__main__":
    # Generate a sample dataset from a normal distribution
    np.random.seed(0)  # For reproducibility
    sample = np.random.normal(loc=100, scale=15, size=50)  # Sample with mean 100, std 15, and size 50

    # Define the population mean and standard deviation
    population_mean = 105
    population_std = 15

    # Perform Z-test
    z_statistic, critical_value, decision = z_test(sample, population_mean, population_std)

    # Print results
    print(f"Z-statistic: {z_statistic:.4f}")
    print(f"Critical value: {critical_value:.4f}")
    print(f"Decision: {decision}")
```

```python
# Implement a one-sample Z-test using Python to compare the sample mean with the population mean.

import numpy as np
from scipy.stats import norm

def z_test(sample, population_mean, population_std, alpha=0.05):
    """
    Perform a one-sample Z-test to compare the sample mean to a known population mean.

    Parameters:
    sample (array-like): The sample data.
    population_mean (float): The population mean.
    population_std (float): The known population standard deviation.
    alpha (float): The significance level (default is 0.05).

    Returns:
    tuple: Z-statistic, critical value, P-value, decision (Reject or Fail to Reject H0)
    """
    # Calculate sample mean and sample size
    sample_mean = np.mean(sample)
    n = len(sample)

    # Calculate the Z-statistic
    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    # Find the critical Z-value for a two-tailed test at the given significance level
    critical_value = norm.ppf(1 - alpha / 2)  # Two-tailed test

    # Calculate the P-value
    p_value = 2 * (1 - norm.cdf(abs(z_statistic)))  # Two-tailed P-value

    # Compare Z-statistic to critical value and make the decision
    if abs(z_statistic) > critical_value:
        decision = "Reject the null hypothesis (H0)"
    else:
        decision = "Fail to reject the null hypothesis (H0)"

    return z_statistic, critical_value, p_value, decision

# Example usage
if __name__ == "__main__":
    # Simulate a sample dataset from a normal distribution (mean = 100, std = 15)
    np.random.seed(0)  # For reproducibility
    sample = np.random.normal(loc=100, scale=15, size=50)  # Sample with mean 100, std 15, and size 50

    # Define the population mean and population standard deviation
    population_mean = 105
    population_std = 15

    # Perform Z-test
    z_statistic, critical_value, p_value, decision = z_test(sample, population_mean, population_std)

    # Print results
    print(f"Z-statistic: {z_statistic:.4f}")
    print(f"Critical value: {critical_value:.4f}")
    print(f"P-value: {p_value:.4f}")
    print(f"Decision: {decision}")
```

```python
#  Perform a two-tailed Z-test using Python and visualize the decision region on a plot

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def z_test(sample, population_mean, population_std, alpha=0.05):
    """
    Perform a two-tailed Z-test for comparing a sample mean to a known population mean.

    Parameters:
    sample (array-like): The sample data.
    population_mean (float): The population mean.
    population_std (float): The known population standard deviation.
    alpha (float): The significance level (default is 0.05).

    Returns:
    tuple: Z-statistic, critical value, P-value, decision (Reject or Fail to Reject H0)
    """
    # Calculate sample mean and sample size
    sample_mean = np.mean(sample)
    n = len(sample)

    # Calculate the Z-statistic
    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    # Find the critical Z-value for a two-tailed test at the given significance level
    critical_value = norm.ppf(1 - alpha / 2)  # Two-tailed test

    # Calculate the P-value
    p_value = 2 * (1 - norm.cdf(abs(z_statistic)))  # Two-tailed P-value

    # Compare Z-statistic to critical value and make the decision
    if abs(z_statistic) > critical_value:
        decision = "Reject the null hypothesis (H0)"
    else:
        decision = "Fail to reject the null hypothesis (H0)"

    return z_statistic, critical_value, p_value, decision

def plot_z_test(z_statistic, critical_value):
    """
    Plot the standard normal distribution and highlight the critical regions for the two-tailed Z-test.

    Parameters:
    z_statistic (float): The calculated Z-statistic.
    critical_value (float): The critical value for the two-tailed test.
    """
    # Create a range of Z values for plotting the standard normal distribution
    z_values = np.linspace(-4, 4, 1000)
    pdf_values = norm.pdf(z_values)

    # Plot the standard normal distribution curve
    plt.plot(z_values, pdf_values, label="Standard Normal Distribution", color="blue")

    # Shade the critical regions (tails)
    plt.fill_between(z_values, 0, pdf_values, where=(z_values <= -critical_value) | (z_values >= critical_value),
                     color="red", alpha=0.5, label="Critical Region (Reject H0)")

    # Mark the calculated Z-statistic on the plot
    plt.axvline(z_statistic, color='green', linestyle='--', label=f"Z-statistic = {z_statistic:.2f}")

    # Add labels and title
    plt.title("Two-Tailed Z-Test: Decision Region")
    plt.xlabel("Z-Value")
    plt.ylabel("Probability Density")
    plt.legend(loc="best")

    # Show the plot
    plt.grid(True)
    plt.show()

# Example usage
if __name__ == "__main__":
    # Simulate a sample dataset from a normal distribution (mean = 100, std = 15)
    np.random.seed(0)  # For reproducibility
    sample = np.random.normal(loc=100, scale=15, size=50)  # Sample with mean 100, std 15, and size 50

    # Define the population mean and population standard deviation
    population_mean = 105
    population_std = 15

    # Perform Z-test
    z_statistic, critical_value, p_value, decision = z_test(sample, population_mean, population_std)

    # Print results
    print(f"Z-statistic: {z_statistic:.4f}")
    print(f"Critical value: {critical_value:.4f}")
    print(f"P-value: {p_value:.4f}")
    print(f"Decision: {decision}")

    # Plot the decision region
    plot_z_test(z_statistic, critical_value)
```

```python
# Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_errors(population_mean, alternative_mean, population_std, alpha=0.05, sample_size=30):
    """
    Visualize Type 1 and Type 2 errors during hypothesis testing for a two-tailed Z-test.

    Parameters:
    population_mean (float): The population mean under the null hypothesis.
    alternative_mean (float): The true population mean under the alternative hypothesis.
    population_std (float): The population standard deviation.
    alpha (float): The significance level (default is 0.05).
    sample_size (int): The sample size.
    """
    # Calculate the critical Z-value for the two-tailed test at the given significance level
    critical_value = norm.ppf(1 - alpha / 2)

    # Calculate the standard error of the mean
    standard_error = population_std / np.sqrt(sample_size)

    # Define the Z-distribution for the null hypothesis
    null_distribution = norm(loc=population_mean, scale=standard_error)

    # Define the Z-distribution for the alternative hypothesis
    alternative_distribution = norm(loc=alternative_mean, scale=standard_error)

    # Create a range of Z values for plotting the distributions
    z_values = np.linspace(population_mean - 4 * population_std, population_mean + 4 * population_std, 1000)

    # Plot the null hypothesis distribution
    plt.plot(z_values, null_distribution.pdf(z_values), label="Null Hypothesis Distribution", color='blue')

    # Plot the alternative hypothesis distribution
    plt.plot(z_values, alternative_distribution.pdf(z_values), label="Alternative Hypothesis Distribution", color='green', linestyle='dashed')

    # Highlight the critical region (rejecting H0)
    plt.fill_between(z_values, 0, null_distribution.pdf(z_values), where=(z_values <= -critical_value) | (z_values >= critical_value),
                     color="red", alpha=0.5, label="Critical Region (Reject H0)")

    # Highlight the Type 1 error region (rejecting H0 when it is true)
    plt.fill_between(z_values, 0, null_distribution.pdf(z_values), where=(z_values <= -critical_value),
                     color="pink", alpha=0.5, label="Type 1 Error Region (False Positive)")
    plt.fill_between(z_values, 0, null_distribution.pdf(z_values), where=(z_values >= critical_value),
                     color="pink", alpha=0.5)

    # Highlight the Type 2 error region (failing to reject H0 when it is false)
    plt.fill_between(z_values, 0, alternative_distribution.pdf(z_values), where=(z_values >= -critical_value) & (z_values <= critical_value),
                     color="yellow", alpha=0.5, label="Type 2 Error Region (False Negative)")

    # Mark the critical value boundaries
    plt.axvline(x=-critical_value, color='red', linestyle='--', label=f'Critical value = {-critical_value:.2f}')
    plt.axvline(x=critical_value, color='red', linestyle='--')

    # Add labels and title
    plt.title("Visualization of Type 1 and Type 2 Errors in Hypothesis Testing")
    plt.xlabel("Z-Value")
    plt.ylabel("Probability Density")
    plt.legend(loc="upper left")
    plt.grid(True)

    # Show the plot
    plt.show()

# Example usage
if __name__ == "__main__":
    # Parameters
    population_mean = 100  # Null hypothesis mean
    alternative_mean = 105  # Alternative hypothesis mean
    population_std = 15  # Population standard deviation
    sample_size = 30  # Sample size

    # Visualize Type 1 and Type 2 errors
    visualize_errors(population_mean, alternative_mean, population_std, alpha=0.05, sample_size=sample_size)
```

```python
# Write a Python program to perform an independent T-test and interpret the results

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def independent_t_test(group1, group2, alpha=0.05):
    """
    Perform an independent T-test to compare the means of two independent groups.

    Parameters:
    group1 (array-like): The first sample data.
    group2 (array-like): The second sample data.
    alpha (float): The significance level (default is 0.05).

    Returns:
    tuple: T-statistic, P-value, degrees of freedom, decision (Reject or Fail to Reject H0)
    """
    # Calculate the means of both groups
    mean1 = np.mean(group1)
    mean2 = np.mean(group2)

    # Calculate the sample standard deviations of both groups
    std1 = np.std(group1, ddof=1)
    std2 = np.std(group2, ddof=1)

    # Sample sizes
    n1 = len(group1)
    n2 = len(group2)

    # Calculate the pooled standard deviation
    pooled_std = np.sqrt(((n1 - 1) * std1**2 + (n2 - 1) * std2**2) / (n1 + n2 - 2))

    # Calculate the t-statistic
    t_statistic = (mean1 - mean2) / (pooled_std * np.sqrt(1/n1 + 1/n2))

    # Calculate the degrees of freedom
    df = n1 + n2 - 2

    # Calculate the P-value using the t-distribution
    p_value = 2 * (1 - stats.t.cdf(abs(t_statistic), df))  # Two-tailed test

    # Compare the P-value with alpha to make the decision
    if p_value <= alpha:
        decision = "Reject the null hypothesis (H0)"
    else:
        decision = "Fail to reject the null hypothesis (H0)"

    return t_statistic, p_value, df, decision

def plot_t_test(group1, group2, t_statistic, alpha=0.05):
    """
    Plot the T-distribution and visualize the T-statistic and critical regions.

    Parameters:
    group1 (array-like): The first sample data.
    group2 (array-like): The second sample data.
    t_statistic (float): The calculated T-statistic.
    alpha (float): The significance level (default is 0.05).
    """
    # Calculate the degrees of freedom
    df = len(group1) + len(group2) - 2

    # Create a range of T values for plotting the T-distribution
    t_values = np.linspace(-5, 5, 1000)
    t_pdf = stats.t.pdf(t_values, df)

    # Plot the T-distribution curve
    plt.plot(t_values, t_pdf, label="T-distribution", color="blue")

    # Mark the critical regions (rejection regions)
    critical_value = stats.t.ppf(1 - alpha / 2, df)  # Two-tailed test critical value
    plt.fill_between(t_values, 0, t_pdf, where=(t_values <= -critical_value) | (t_values >= critical_value),
                     color="red", alpha=0.5, label="Critical Region (Reject H0)")

    # Mark the calculated T-statistic on the plot
    plt.axvline(x=t_statistic, color='green', linestyle='--', label=f"T-statistic = {t_statistic:.2f}")

    # Add labels and title
    plt.title("Independent T-Test: Decision Region")
    plt.xlabel("T-Value")
    plt.ylabel("Probability Density")
    plt.legend(loc="best")

    # Show the plot
    plt.grid(True)
    plt.show()

# Example usage
if __name__ == "__main__":
    # Simulate data for two groups
    np.random.seed(0)  # For reproducibility
    group1 = np.random.normal(loc=50, scale=10, size=30)  # Group 1 (mean=50, std=10)
    group2 = np.random.normal(loc=55, scale=12, size=30)  # Group 2 (mean=55, std=12)

    # Perform Independent T-test
    t_statistic, p_value, df, decision = independent_t_test(group1, group2, alpha=0.05)

    # Print the results
    print(f"T-statistic: {t_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")
    print(f"Degrees of freedom: {df}")
    print(f"Decision: {decision}")

    # Plot the T-test results
    plot_t_test(group1, group2, t_statistic)
```

```python
# Perform a paired sample T-test using Python and visualize the comparison results

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def paired_t_test(before, after, alpha=0.05):
    """
    Perform a paired T-test to compare the means of two related groups (before and after).

    Parameters:
    before (array-like): The values from the first group (e.g., before treatment).
    after (array-like): The values from the second group (e.g., after treatment).
    alpha (float): The significance level (default is 0.05).

    Returns:
    tuple: T-statistic, P-value, degrees of freedom, decision (Reject or Fail to Reject H0)
    """
    # Calculate the differences between the paired observations
    differences = np.array(after) - np.array(before)

    # Calculate the mean and standard deviation of the differences
    mean_diff = np.mean(differences)
    std_diff = np.std(differences, ddof=1)

    # Sample size
    n = len(differences)

    # Calculate the t-statistic
    t_statistic = mean_diff / (std_diff / np.sqrt(n))

    # Calculate the degrees of freedom
    df = n - 1

    # Calculate the P-value for a two-tailed test
    p_value = 2 * (1 - stats.t.cdf(abs(t_statistic), df))

    # Compare the P-value with alpha to make the decision
    if p_value <= alpha:
        decision = "Reject the null hypothesis (H0)"
    else:
        decision = "Fail to reject the null hypothesis (H0)"

    return t_statistic, p_value, df, decision, differences

def plot_paired_comparison(before, after, t_statistic, alpha=0.05):
    """
    Visualize the paired comparison of the two groups (before and after) along with the T-test result.

    Parameters:
    before (array-like): The values from the first group (e.g., before treatment).
    after (array-like): The values from the second group (e.g., after treatment).
    t_statistic (float): The calculated T-statistic.
    alpha (float): The significance level (default is 0.05).
    """
    # Create a figure for visualization
    plt.figure(figsize=(8, 6))

    # Plot the before and after measurements
    plt.plot(before, 'bo', label="Before", alpha=0.7)
    plt.plot(after, 'ro', label="After", alpha=0.7)

    # Connect the paired data points with lines
    for i in range(len(before)):
        plt.plot([i, i], [before[i], after[i]], 'k-', alpha=0.5)

    # Add the T-statistic line (mean difference)
    plt.axhline(y=np.mean(after) - np.mean(before), color='green', linestyle='--', label=f"Mean Difference = {np.mean(after) - np.mean(before):.2f}")

    # Add labels and title
    plt.title("Paired T-Test: Comparison of Before and After Measurements")
    plt.xlabel("Paired Observations")
    plt.ylabel("Values")
    plt.legend()

    # Show the plot
    plt.grid(True)
    plt.show()

# Example usage
if __name__ == "__main__":
    # Simulate data for before and after measurements (e.g., before and after treatment)
    np.random.seed(0)  # For reproducibility
    before = np.random.normal(loc=50, scale=10, size=30)  # Before treatment
    after = before + np.random.normal(loc=5, scale=5, size=30)  # After treatment (with a slight increase)

    # Perform Paired T-test
    t_statistic, p_value, df, decision, differences = paired_t_test(before, after, alpha=0.05)

    # Print the results
    print(f"T-statistic: {t_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")
    print(f"Degrees of freedom: {df}")
    print(f"Decision: {decision}")

    # Plot the results of the paired comparison
    plot_paired_comparison(before, after, t_statistic)
```

```python
# Simulate data and perform both Z-test and T-test, then compare the results using Python


import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def z_test(sample, population_mean, population_std, alpha=0.05):
    """
    Perform a Z-test to compare the sample mean to the population mean.

    Parameters:
    sample (array-like): The sample data.
    population_mean (float): The population mean.
    population_std (float): The population standard deviation.
    alpha (float): The significance level (default is 0.05).

    Returns:
    tuple: Z-statistic, P-value, decision (Reject or Fail to Reject H0)
    """
    # Sample mean and size
    sample_mean = np.mean(sample)
    n = len(sample)

    # Z-statistic calculation
    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    # P-value for two-tailed test
    p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

    # Decision based on the P-value
    if p_value <= alpha:
        decision = "Reject the null hypothesis (H0)"
    else:
        decision = "Fail to reject the null hypothesis (H0)"

    return z_statistic, p_value, decision

def t_test(sample, population_mean, alpha=0.05):
    """
    Perform a T-test to compare the sample mean to the population mean.

    Parameters:
    sample (array-like): The sample data.
    population_mean (float): The population mean.
    alpha (float): The significance level (default is 0.05).

    Returns:
    tuple: T-statistic, P-value, decision (Reject or Fail to Reject H0)
    """
    # Sample mean, sample standard deviation, and sample size
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)
    n = len(sample)

    # T-statistic calculation
    t_statistic = (sample_mean - population_mean) / (sample_std / np.sqrt(n))

    # Degrees of freedom
    df = n - 1

    # P-value for two-tailed test
    p_value = 2 * (1 - stats.t.cdf(abs(t_statistic), df))

    # Decision based on the P-value
    if p_value <= alpha:
        decision = "Reject the null hypothesis (H0)"
    else:
        decision = "Fail to reject the null hypothesis (H0)"

    return t_statistic, p_value, decision

def compare_z_t_tests():
    # Simulate data from a normal distribution
    np.random.seed(0)  # For reproducibility
    sample = np.random.normal(loc=50, scale=10, size=30)  # Sample of size 30, mean=50, std=10

    population_mean = 50  # Assume the population mean is 50
    population_std = 10   # Assume the population standard deviation is 10

    # Perform Z-test
    z_statistic, z_p_value, z_decision = z_test(sample, population_mean, population_std)

    # Perform T-test
    t_statistic, t_p_value, t_decision = t_test(sample, population_mean)

    # Print the results of both tests
    print(f"Z-test result:")
    print(f"Z-statistic: {z_statistic:.4f}")
    print(f"P-value: {z_p_value:.4f}")
    print(f"Decision: {z_decision}")

    print("\nT-test result:")
    print(f"T-statistic: {t_statistic:.4f}")
    print(f"P-value: {t_p_value:.4f}")
    print(f"Decision: {t_decision}")

    # Visualize the comparison results
    plt.figure(figsize=(8, 6))

    # Plot sample data distribution
    plt.hist(sample, bins=10, alpha=0.7, color='blue', label="Sample Data")

    # Plot population mean as a vertical line
    plt.axvline(population_mean, color='red', linestyle='dashed', label=f"Population Mean = {population_mean}")

    # Add the Z-statistic and T-statistic lines
    plt.axvline(np.mean(sample) + z_statistic * (population_std / np.sqrt(len(sample))), color='green', linestyle='-', label=f"Z-test Statistic = {z_statistic:.2f}")
    plt.axvline(np.mean(sample) + t_statistic * (np.std(sample, ddof=1) / np.sqrt(len(sample))), color='purple', linestyle='-', label=f"T-test Statistic = {t_statistic:.2f}")

    # Add labels and title
    plt.title("Z-test vs T-test Comparison")
    plt.xlabel("Sample Values")
    plt.ylabel("Frequency")
    plt.legend()

    # Show the plot
    plt.grid(True)
    plt.show()

# Run the comparison
compare_z_t_tests()
```

```python
# Write a Python function to calculate the confidence interval for a sample mean and explain its significance

import numpy as np
import scipy.stats as stats

def confidence_interval(sample, confidence_level=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
    sample (array-like): The sample data.
    confidence_level (float): The confidence level (default is 0.95).

    Returns:
    tuple: Lower bound, upper bound, sample mean, and sample standard deviation.
    """
    # Sample mean and standard deviation
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)

    # Sample size
    n = len(sample)

    # Degrees of freedom
    df = n - 1

    # Find the t-statistic for the given confidence level
    t_statistic = stats.t.ppf(1 - (1 - confidence_level) / 2, df)

    # Calculate the margin of error
    margin_of_error = t_statistic * (sample_std / np.sqrt(n))

    # Calculate the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound, sample_mean, sample_std

# Example usage
if __name__ == "__main__":
    # Simulate data for a sample (e.g., test scores of 30 students)
    np.random.seed(42)
    sample = np.random.normal(loc=50, scale=10, size=30)  # Mean = 50, SD = 10, Sample size = 30

    # Calculate the confidence interval for the sample mean
    lower, upper, mean, std = confidence_interval(sample, confidence_level=0.95)

    # Print the results
    print(f"Sample Mean: {mean:.2f}")
    print(f"Sample Standard Deviation: {std:.2f}")
    print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")
```

```python
# Write a Python program to calculate the margin of error for a given confidence level using sample data

import numpy as np
import scipy.stats as stats

def margin_of_error(sample, confidence_level=0.95):
    """
    Calculate the margin of error for a given sample and confidence level.

    Parameters:
    sample (array-like): The sample data.
    confidence_level (float): The confidence level (default is 0.95).

    Returns:
    float: The margin of error.
    """
    # Sample size
    n = len(sample)

    # Sample standard deviation (using ddof=1 for sample standard deviation)
    sample_std = np.std(sample, ddof=1)

    # Degrees of freedom
    df = n - 1

    # Critical t-value from the t-distribution for the given confidence level
    t_statistic = stats.t.ppf(1 - (1 - confidence_level) / 2, df)

    # Standard error of the mean
    standard_error = sample_std / np.sqrt(n)

    # Calculate the margin of error
    margin_of_error = t_statistic * standard_error

    return margin_of_error

# Example usage
if __name__ == "__main__":
    # Simulate data for a sample (e.g., test scores of 30 students)
    np.random.seed(42)
    sample_data = np.random.normal(loc=50, scale=10, size=30)  # Mean = 50, SD = 10, Sample size = 30

    # Calculate the margin of error for a 95% confidence level
    margin_error = margin_of_error(sample_data, confidence_level=0.95)

    # Print the result
    print(f"Sample Mean: {np.mean(sample_data):.2f}")
    print(f"Sample Standard Deviation: {np.std(sample_data, ddof=1):.2f}")
    print(f"Margin of Error at 95% Confidence Level: {margin_error:.2f}")
```

```python
# Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process


import numpy as np

# Bayes' Theorem: P(H|E) = P(E|H) * P(H) / P(E)

def bayesian_inference(prior, likelihood, evidence):
    """
    Perform Bayesian inference using Bayes' Theorem.

    Parameters:
    prior (list): Prior probabilities for the hypothesis.
    likelihood (list): Likelihood of observing the evidence for each hypothesis.
    evidence (float): Total probability of the evidence.

    Returns:
    list: Posterior probabilities for each hypothesis.
    """
    # Calculate the posterior using Bayes' Theorem
    posterior = [(likelihood[i] * prior[i]) / evidence for i in range(len(prior))]

    return posterior

def main():
    # Given data
    heads = 8
    tosses = 10
    evidence = 0.5  # Probability of heads for the fair coin

    # Hypotheses: Coin 1 (Fair coin) and Coin 2 (Biased coin)
    prior_fair = 0.5  # Prior probability of choosing the fair coin
    prior_biased = 0.5  # Prior probability of choosing the biased coin

    # Likelihood: P(E|H) for each hypothesis (fair coin vs biased coin)
    # For fair coin: P(E|H) is the binomial likelihood of getting 'heads' times 'heads' out of 'tosses'
    # For biased coin: P(E|H) is the binomial likelihood using biased coin's head probability (0.7)

    # Binomial likelihood function (probability of getting exactly 'heads' heads in 'tosses' tosses)
    def binomial_likelihood(p_head, heads, tosses):
        return (np.math.comb(tosses, heads)) * (p_head**heads) * ((1 - p_head)**(tosses - heads))

    # Likelihood for each hypothesis
    likelihood_fair = binomial_likelihood(0.5, heads, tosses)
    likelihood_biased = binomial_likelihood(0.7, heads, tosses)

    # Evidence: Total probability of observing the evidence (marginal likelihood)
    # P(E) = P(E|H1)*P(H1) + P(E|H2)*P(H2)
    evidence = (likelihood_fair * prior_fair) + (likelihood_biased * prior_biased)

    # Prior probabilities for Coin 1 (fair) and Coin 2 (biased)
    prior = [prior_fair, prior_biased]

    # Likelihoods for the two coins
    likelihood = [likelihood_fair, likelihood_biased]

    # Perform Bayesian Inference
    posterior = bayesian_inference(prior, likelihood, evidence)

    # Output the results
    print(f"Posterior Probability of Fair Coin: {posterior[0]:.4f}")
    print(f"Posterior Probability of Biased Coin: {posterior[1]:.4f}")

if __name__ == "__main__":
    main()
```

```python
# Perform a Chi-square test for independence between two categorical variables in Python


import numpy as np
import pandas as pd
import scipy.stats as stats

# Example data: Preference for tea or coffee by gender (Male/Female)
data = {
    'Tea': [30, 10],  # Tea preference (Male, Female)
    'Coffee': [20, 40]  # Coffee preference (Male, Female)
}

# Convert data into a DataFrame
df = pd.DataFrame(data, index=['Male', 'Female'])

# Perform the Chi-square test for independence
chi2_stat, p_value, dof, expected = stats.chi2_contingency(df)

# Output the results
print(f"Chi-Square Statistic: {chi2_stat:.4f}")
print(f"P-Value: {p_value:.4f}")
print(f"Degrees of Freedom: {dof}")
print(f"Expected Frequencies: \n{expected}")

# Interpretation
alpha = 0.05  # significance level

if p_value < alpha:
    print("Reject the null hypothesis: There is an association between gender and preference.")
else:
    print("Fail to reject the null hypothesis: There is no association between gender and preference.")
```

```python
''' Write a Python program to calculate the expected frequencies for a Chi-square test based on observed
data '''


import numpy as np

def calculate_expected_frequencies(observed):
    """
    Calculate the expected frequencies for a Chi-square test based on observed data.

    Parameters:
    observed (numpy array): 2D array representing the contingency table of observed frequencies.

    Returns:
    numpy array: 2D array of expected frequencies.
    """
    # Total number of observations (sum of all observed frequencies)
    total = np.sum(observed)

    # Row totals
    row_totals = np.sum(observed, axis=1)

    # Column totals
    col_totals = np.sum(observed, axis=0)

    # Calculate expected frequencies
    expected = np.outer(row_totals, col_totals) / total

    return expected

# Example observed data (2x2 contingency table)
observed_data = np.array([[30, 10],  # 30 males prefer tea, 10 females prefer tea
                          [20, 40]])  # 20 males prefer coffee, 40 females prefer coffee

# Calculate expected frequencies
expected_data = calculate_expected_frequencies(observed_data)

# Display results
print("Observed Frequencies:")
print(observed_data)
print("\nExpected Frequencies:")
print(expected_data)
```

```python
# Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution

import numpy as np
import scipy.stats as stats

# Example: Observed frequencies of rolling a 6-sided die
observed = np.array([18, 22, 19, 21, 20, 20])  # Observed frequencies for each face of the die (1, 2, 3, 4, 5, 6)

# Expected frequencies for a fair die (uniform distribution)
n = np.sum(observed)  # Total number of rolls
expected = np.array([n / 6] * 6)  # Each side of the die should have the same expected frequency if fair

# Perform the Chi-square goodness-of-fit test
chi2_stat, p_value = stats.chisquare(observed, expected)

# Output the results
print(f"Chi-Square Statistic: {chi2_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

# Interpretation of the result
alpha = 0.05  # Significance level

if p_value < alpha:
    print("Reject the null hypothesis: The die is not fair.")
else:
    print("Fail to reject the null hypothesis: The die is fair.")
```
